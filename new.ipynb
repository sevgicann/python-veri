```python

base_url = "https://turkishnetworktimes.com/kategori/gundem/page/{}"
header_list = []

for page_number in range(1, 51):  # İlk 50 sayfaya göre dön
    url = base_url.format(page_number)
    response = requests.get(url)
    
    if response.status_code == 200:
        bsobj = soup(response.content, 'html.parser')
        for news in bsobj.findAll(['h2', 'strong']):
            header_list.append(news.text.strip())

header_list
```

```python

base_url = "https://turkishnetworktimes.com/kategori/gundem/page/{}"
url_list = []

for page_number in range(1, 51):  # İlk 50 sayfaya göre dön
    url = base_url.format(page_number)
    response = requests.get(url)
    
    if response.status_code == 200:
        bsobj = soup(response.content, 'lxml')
        for news in bsobj.findAll('a', {'class': 'post-link'}):
            url_list.append(news['href'])

url_list
```

```python
summary=[]
for link in url:
    page=requests.get(link)
    bsobj=soup(page.content)
    for news in bsobj.findAll('div',{'class':'haber-desc'}):
        text.append(news.text.strip())
summary
```

```python
update_date = []

for link in url:
    page = requests.get(link)
    bsobj = soup(page.content, 'html.parser')
    
    for news in bsobj.findAll('span', {'class': 'tarih'}):
        if 'Güncelleme:' in news.get_text():
            update_date.append(news.get_text().strip())

update_date
```

```python
publish_date = []

for link in url:
    page = requests.get(link)
    bsobj = soup(page.content, 'html.parser')
    
    for news in bsobj.findAll('span', {'class': 'tarih'}):
        if 'Yayınlanma:' in news.get_text():
            publish_date.append(news.get_text().strip())

publish_date
```

```python
img_url_list=[]
for news in b.findAll('img',{'class':'wp-post-image'}):
    img_url_list.append(news['src'])
img_url_list
```

```python
text=[]
for link in url:
    page=requests.get(link)
    bsobj=soup(page.content)
    for news in bsobj.findAll('div',{'class':'yazi_icerik'}):
        text.append(news.text.strip())
text
```

```python
import csv



# CSV dosyası adı ve yazma modu
file_name = "verilerde.csv"

# Veri başlıkları
column_names = ['URL', 'Header', 'Summary', 'Text', 'Img URL', 'Publish Date','Update Date']

# Verileri birleştirme
combined_data = zip(url, header, summary, text, img_url_list, publish_date, update_date)

# CSV dosyasına yazma
with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Başlık satırını yazma
    writer.writerow(column_names)

    # Her bir satırı CSV dosyasına yazma
    for row in combined_data:
        writer.writerow(row)

print(f"{file_name} adlı CSV dosyası oluşturuldu.")
```

```python
from collections import Counter
from matplotlib import pyplot as plt

# text listesindeki metinleri birleştirme
all_text = ' '.join(text)

# Kelimeleri ayıklama ve küçük harfe dönüştürme
words = all_text.lower().split()

# Kelime frekanslarını hesaplama
word_freq = Counter(words)

# En çok kullanılan 10 kelimeyi bulma
top_words = word_freq.most_common(10)
top_words, word_freq

# Grafiği oluşturma
top_word, freq = zip(*top_words)
plt.figure(figsize=(10, 6))
plt.bar(top_word, freq, color='skyblue')
plt.title('En Çok Kullanılan Kelimelerin Frekansı')
plt.xlabel('Kelimeler')
plt.ylabel('Frekans')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
