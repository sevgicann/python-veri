```python
r=requests.get("https://turkishnetworktimes.com/kategori/gundem/")
```

```python
b=soup(r.content,'lxml')
```

```python
header=[]
for news in b.findAll(['h2', 'strong']):
  
  #print(news.text)
  header.append(news.text)
header
```

```python
url=[]
for news in b.findAll('a',{'class':'post-link'}):
    url.append(news['href'])
url
```

```python
summary=[]

for news in b.findAll('div',{'class':'haber-desc'}):
  
  #print(news.text)
  summary.append(news.text)
summary
```

```python
publish_date=[]

for news in b.findAll('span',{'class':'tarih'}):
  
  #print(news.text)
  publish_date.append(news.text)
publish_date
```

```python
publish_date = []

for link in url:
    page = requests.get(link)
    bsobj = soup(page.content, 'html.parser')
    
    span_tags = bsobj.findAll('span')  # Sadece 'span' etiketlerini seç
    for tag in span_tags:
        if 'tarih' in tag.get('class', []):  # Eğer 'tarih' class'ına sahipse
            publish_date.append(tag.get_text().strip())

publish_date
```

```python
img_url_list=[]
for news in b.findAll('img',{'class':'wp-post-image'}):
    img_url_list.append(news['src'])
img_url_list
```

```python
text=[]
for link in url:
    page=requests.get(link)
    bsobj=soup(page.content)
    for news in bsobj.findAll('div',{'class':'yazi_icerik'}):
        text.append(news.text.strip())
text
```

```python
import csv



# CSV dosyası adı ve yazma modu
file_name = "verilerden.csv"

# Veri başlıkları
column_names = ['URL', 'Header', 'Summary', 'Text', 'Img URL', 'Publish Date']

# Verileri birleştirme
combined_data = zip(url, header, summary, text, img_url_list, publish_date)

# CSV dosyasına yazma
with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Başlık satırını yazma
    writer.writerow(column_names)

    # Her bir satırı CSV dosyasına yazma
    for row in combined_data:
        writer.writerow(row)

print(f"{file_name} adlı CSV dosyası oluşturuldu.")
```
