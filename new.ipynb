```python
r=requests.get("https://turkishnetworktimes.com/kategori/gundem/")
```

```python
b=soup(r.content,'lxml')
```

```python
header=[]
for news in b.findAll(['h2', 'strong']):
  
  #print(news.text)
  header.append(news.text)
header
```

```python
links=[]
for news in b.findAll('a',{'class':'post-link'}):
    links.append(news['href'])
links
```

```python
text=[]
for link in links:
    page=requests.get(link)
    bsobj=soup(page.content)
    for news in bsobj.findAll('div',{'class':'yazi_icerik'}):
        text.append(news.text.strip())
text
```

```python
import csv



# Dosya adı ve açma modunu belirleyin (w -> write)
file_name = "verileri.csv"

# En kısa listenin uzunluğunu bulun
min_length = min(len(text), len(links), len(header))

# CSV dosyasını yazma modunda açın ve writer objesini oluşturun
with open(file_name, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    # İlk olarak başlık satırını (header) yazın
    writer.writerow(['Text', 'Link', 'Header'])

    # En kısa listenin uzunluğu kadar olan elemanları CSV dosyasına yazın
    for i in range(min_length):
        writer.writerow([text[i], links[i], header[i]])

print(f"{file_name} adlı CSV dosyası oluşturuldu.")
```
